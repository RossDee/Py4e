```python
df_population_raw = pd.read_csv('population_total.csv')
df_population_raw
```

```python
# reshape the data ,droping null value
# inplace attibute to update the data frame
df_population_raw.dropna(inplace=True)
```

```python
# year as index, country as column, population as data 
df_pivot=df_population_raw.pivot(
    index = 'year',
    columns = 'country',
    values = 'population'
    )
```

```python
#
df_pivot=df_pivot[['Albania','United States','India','Uruguay']]
df_pivot
```

```python
# lineplot
df_pivot.plot(kind='line',xlabel='Year',ylabel='Population')
```

```python
# index attibute, isin to find the index value
df_pivot_2020 = df_pivot[df_pivot.index.isin([2020])]
df_pivot_2020
```

```python
# transpose dataframe to get the index for bar chart
df_pivot_2020.T
```

```python
df_pivot_2020= df_pivot_2020.T
```

```python
df_pivot_2020.plot(kind ='bar')
```

```python
df_pivot_multiyears = df_pivot[df_pivot.index.isin([1980,1990,2000,2010,2020])]
```

```python
df_pivot_multiyears.plot(kind ='bar')
```

```python
df_pivot_multiyears = df_pivot_multiyears.T
```

```python
df_pivot_multiyears
```

```python
df_pivot_multiyears.plot(kind ='bar')
```

```python
# pie chart
df_pivot_2020.rename(columns={2020:'2020'},inplace=True)
df_pivot_2020.plot(kind='pie',y='2020')
```
